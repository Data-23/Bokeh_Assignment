```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource
import pandas as pd

# Step 1: Prepare your data
data = pd.DataFrame({
    'x': [1, 2, 3, 4, 5],
    'y': [6, 7, 2, 4, 5],
    'size': [10, 20, 30, 40, 50],
    'color': ['navy', 'orange', 'green', 'red', 'purple']
})

source = ColumnDataSource(data)

# Step 2: Create a figure
p = figure(title="Simple Bokeh Scatter Plot", x_axis_label='X Axis', y_axis_label='Y Axis')

# Step 3: Add glyphs
p.circle('x', 'y', size='size', color='color', alpha=0.5, source=source)

# Step 4: Customize the plot (optional)
p.background_fill_color = "lightgrey"
p.grid.grid_line_color = "white"

# Step 5: Show the plot
output_notebook()  # Use output_file('plot.html') to save as an HTML file
show(p)
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Step 1: Prepare your data
data = {
    'x': [1, 2, 3, 4, 5],
    'y': [6, 7, 2, 4, 5],
    'size': [10, 20, 30, 40, 50],
    'color': ['navy', 'orange', 'green', 'red', 'purple']
}

source = ColumnDataSource(data)

# Step 2: Create a figure
p = figure(title="Bokeh Plot with Circles and Lines", x_axis_label='X Axis', y_axis_label='Y Axis')

# Step 3: Add glyphs
# Adding circles
p.circle('x', 'y', size='size', color='color', alpha=0.6, source=source, legend_label='Circles')

# Adding lines
p.line('x', 'y', line_width=2, line_color='black', source=source, legend_label='Line')

# Step 4: Customize the plot (optional)
p.background_fill_color = "lightgrey"
p.grid.grid_line_color = "white"
p.legend.location = "top_left"

# Step 5: Show the plot
output_notebook()  # Use output_file('plot.html') to save as an HTML file
show(p)
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Step 1: Prepare your data
data = {
    'x': [1, 2, 3, 4, 5],
    'y': [6, 7, 2, 4, 5],
    'size': [10, 20, 30, 40, 50],
    'color': ['navy', 'orange', 'green', 'red', 'purple']
}

source = ColumnDataSource(data)

# Step 2: Create a figure
p = figure(title="Customized Bokeh Plot", x_axis_label='X Axis', y_axis_label='Y Axis')

# Step 3: Add glyphs
p.circle('x', 'y', size='size', color='color', alpha=0.6, source=source, legend_label='Data Points')

# Step 4: Customize the plot

# Title customization
p.title.text = "Customized Bokeh Plot"
p.title.align = "center"
p.title.text_color = "black"
p.title.text_font_size = "20pt"

# X-axis customization
p.xaxis.axis_label = "X Axis"
p.xaxis.axis_label_text_color = "blue"
p.xaxis.axis_label_text_font_size = "14pt"
p.xaxis.major_label_text_color = "blue"
p.xaxis.major_label_text_font_size = "12pt"
p.xaxis.axis_line_width = 2
p.xaxis.axis_line_color = "blue"
p.xaxis.minor_tick_line_color = "blue"

# Y-axis customization
p.yaxis.axis_label = "Y Axis"
p.yaxis.axis_label_text_color = "green"
p.yaxis.axis_label_text_font_size = "14pt"
p.yaxis.major_label_text_color = "green"
p.yaxis.major_label_text_font_size = "12pt"
p.yaxis.axis_line_width = 2
p.yaxis.axis_line_color = "green"
p.yaxis.minor_tick_line_color = "green"

# Grid customization
p.grid.grid_line_dash = [6, 4]
p.grid.grid_line_color = "gray"

# Legend customization
p.legend.title = "Legend"
p.legend.title_text_font_size = "14pt"
p.legend.label_text_font_size = "12pt"
p.legend.label_text_color = "black"
p.legend.background_fill_color = "lightgray"
p.legend.border_line_color = "black"
p.legend.border_line_width = 1
p.legend.location = "top_left"

# Background customization
p.background_fill_color = "lightyellow"

# Step 5: Show the plot
output_notebook()  # Use output_file('customized_plot.html') to save as an HTML file
show(p)
```

```python
from bokeh.io import curdoc
from bokeh.layouts import column
from bokeh.models import ColumnDataSource, Slider
from bokeh.plotting import figure

# Prepare the data
x = list(range(10))
y = [i**2 for i in x]
source = ColumnDataSource(data=dict(x=x, y=y))

# Create a figure
p = figure(title="Interactive Bokeh Plot", x_axis_label='X', y_axis_label='Y')

# Add a line renderer
line = p.line('x', 'y', source=source, line_width=2, line_color='blue')

# Define a callback function to update the data
def update_data(attr, old, new):
    factor = slider.value
    new_y = [i**factor for i in x]
    source.data = dict(x=x, y=new_y)

# Create a slider widget
slider = Slider(start=0.1, end=4, value=2, step=0.1, title="Power")

# Attach the callback to the slider
slider.on_change('value', update_data)

# Layout the plot and the widget
layout = column(p, slider)

# Add the layout to the current document
curdoc().add_root(layout)
curdoc().title = "Interactive Bokeh Application"


```
